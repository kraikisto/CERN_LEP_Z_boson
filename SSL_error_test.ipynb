```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


dataset_alt = pd.read_csv('https://opendata.cern.ch/record/5201/files/Dimuon_DoubleMu.csv',index_col=False)
dataset_alt.head()
```

```python
lowerlimit = 0
upperlimit = 150
bins = 100


fig = plt.figure(figsize=(15, 10))
plt.hist(dataset_alt.M, bins=bins, range=(lowerlimit, upperlimit))

plt.xlabel('Invariant mass [GeV/cÂ²]', fontsize=15)
plt.ylabel('Number of events', fontsize=15)
plt.title('Invariant mass', fontsize=15) 

plt.show()
```
